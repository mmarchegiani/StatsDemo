```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2
```

```python
chi2
```

## Plot chi-square p.d.f.
We plot the p.d.f. of the chi-2 distribution, for a given number of degrees of freedom (d.o.f.). 

```python
dof = 7

x = np.linspace(0, 100, 101)
plt.plot(x, chi2.pdf(x, dof))
plt.title(f"Chi-2 p.d.f. for {dof} degrees of freedom");
```

## Compute the chi-2 probability
In order to compute the chi-2 probability, we use the c.d.f. of the chi-2 distribution. The output of the c.d.f. is the integral of the p.d.f. up to a certain value. To get the probability for a measured chi-2 $\chi_0^2$ to be greater than a certain value, we just need to compute:

$prob(\chi^2 \geq \chi_0^2, dof) = 1 - cdf(\chi_0^2, dof) $

```python
p = chi2.cdf(20, dof)
p
```

```python
prob_chi2 = 1-p
prob_chi2
```

## Compare chi-2 distributions for different dof
We plot the chi-2 p.d.f. for different values of degrees of freedom.
For increasing d.o.f, the distribution is shifted to greater values, and its absolute peak is reduced.

```python
for dof in [10, 20, 30]:
    plt.plot(x, chi2.pdf(x, dof), label=f"dof={dof}")
    plt.legend(loc="best");
```
